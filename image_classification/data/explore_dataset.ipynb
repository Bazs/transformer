```python
import sys
import os
import numpy as np
from pathlib import Path
sys.path.append("~/repos/transformer")

from image_classification.data.dataset import Config, create_train_val_datasets

train_dataset, val_dataset = create_train_val_datasets(config=Config(root_dir=Path("/home/balazsopra/Documents/Debug/image_classification/datasets/cats_vs_dogs"), train_val_ratio=1.0, train_transform=None))

image_sizes = set()
# Compute mean and std of pixel values, and store image sizes
r = []
g = []
b = []
pixel_count = 0
for image, label in train_dataset:
    image_sizes.add(image.shape)
    r.extend(image[:, :, 0].flatten().tolist())
    g.extend(image[:, :, 1].flatten().tolist())
    b.extend(image[:, :, 2].flatten().tolist())
    pixel_count += image.shape[0] * image.shape[1]
```

```python
mean = (np.sum(r) / pixel_count, np.sum(g) / pixel_count, np.sum(b) / pixel_count)
print(mean)
std = np.std(r), np.std(g), np.std(b)
print(std)

print(image_sizes)
```

```python

widths = {image_size[1] for image_size in image_sizes}
heights = {image_size[2] for image_size in image_sizes}
mean_image_size = (sum(widths) / len(widths), sum(heights) / len(heights))
print(f"Mean image size: {mean_image_size}, max width: {max(widths)}, max height: {max(heights)}")
print(f"Min width: {min(widths)}, min height: {min(heights)}")
```
